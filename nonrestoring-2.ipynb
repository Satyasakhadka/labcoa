```python

def AND(num1,num2):
    if(num1=='1' and num2=='1'):
        return '1'
    else:
        return '0'
def OR(num1,num2):
    if(num1=='1' or num2=='1'):
        return '1'
    else:
        return '0'
def XOR(num1,num2):
    if(num1==num2):
        return '0'
    else:
        return '1'

    
```

```python

def SUB(num_1,num_2):
    n=len(num_1)
    ones_comp=''
    for bit in num_2:
        if bit=='0':
            ones_comp+='1'
        else:
            ones_comp+='0'
    twos_comp=bin(int(ones_comp,2)+1)[2:]#slices the first two character 0b
    if (len(twos_comp)<len(num_2)):
        twos_comp='0'*(len(num_2)-len(twos_comp))+twos_comp
    reverse_num1=num_1[::-1]
    reverse_twoscomp=twos_comp[::-1]
    cin='0'
    sum=''
    for i in range(0,n):
        sum = sum + XOR(XOR(reverse_num1[i],reverse_twoscomp[i]),cin)
        intermediate_xor = XOR(reverse_num1[i], reverse_twoscomp[i])
        intermediate_and = AND(reverse_num1[i], reverse_twoscomp[i])
        cin = OR(intermediate_and, AND(intermediate_xor,cin))
    sum=sum[::-1]
    return sum
```

```python
def add(num_1,num_2):
    n=len(num_1)
    
    reverse_num1=num_1[::-1]

    num_2=num_2.zfill(n)
    reverse_num2=num_2[::-1]
    cin='0'
    sum=''
    for i in range(0,n):
        sum = sum + XOR(XOR(reverse_num1[i],reverse_num2[i]),cin)
        intermediate_xor = XOR(reverse_num1[i], reverse_num2[i])
        intermediate_and = AND(reverse_num1[i], reverse_num2[i])
        cin = OR(intermediate_and, AND(intermediate_xor,cin))
    sum=sum[::-1]
    return sum



   
    
```

```python
def restoringdivision(q,m,a):
    count=len(m)
    print('initial values:a:',a,'q:',q,'m:',m)
    while count:
        print("\nstep:", len(m) - count + 1, '  Left Shift and Subtract:\n')
        a = a[1:] + q[0]
        q = q[1:]
        print('a:', a, 'q:', q)
        a = SUB(a, m)
        print('\na:', a)
        if a[0] == '1':
            q = q + '0'
            a = add(a, m)
        else:
            q = q + '1'
        print('\na:', a, 'q:', q)
        count = count - 1

    # Termination condition
        if count == 0:
            break
    print("\nQuotient:",q)
    print("\nRemainder:",a)
```

```python
dividend=input("Enter the value of dividend:")
divisor=input("Enter the value of divisor:")
accumulator='0'*len(dividend)
restoringdivision(dividend,divisor,accumulator)
    
```
